```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
files = [file for file in os.listdir("covid_tweet") if "tag_#" in file]
print(files)
```

```python
import re
def remove_noise(text):
        result = re.sub("http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+","", text)
        result = re.sub("(@[A-Za-z0-9_]+)","", result)
        result = re.sub("\\n"," ", result)
        return result
```

```python
text = "“Unplanned #lockdown is the product of one person’s ego and because of it coronavirus has spread throughout the country,” #RahulGandhi said \n https://t.co/V9wRIDyCmG"
remove_noise(text)
```

```python
import pandas as pd
from datetime import datetime

# nltk.org/api/nltk.sentiment.html
for file in files: 
    print(file)
    # df = pd.read_csv('./covid_tweet/' + file)
    df = pd.read_csv('./covid_tweet/' + file, lineterminator='\n')
    df['Text'] = df['Text'].apply(remove_noise)

    sentiment_added = pd.concat([df, df.Text.apply(lambda row: sid.polarity_scores(row)).apply(pd.Series)], axis= 1)
    sentiment_added = sentiment_added[sentiment_added.columns[1:]]

    # #TODO: data cleaning!!
    sentiment_added["date"] = pd.to_datetime(sentiment_added["date"], errors = "coerce").dt.date
    sentiment_added = sentiment_added.dropna()
    sentiment_added.head()
    sentiment_added.to_csv('./tweet_sentiment_added_noise_removed/' + file)
```

```python
# neutral texts
sentiment_added.sort_values("compound").query("compound > -0.05 & compound < 0.05")["Text"].tail(10).values
```

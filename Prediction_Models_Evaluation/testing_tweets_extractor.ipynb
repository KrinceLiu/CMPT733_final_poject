```python
import os

# the path variables for required folders

# the directory that contains all the sub folders or files
main_folder = '/content/drive/MyDrive/CMPT733_Final_Project'

# the folder that contains the full data of pretended target hashtags which are described in section 5.2 in the report
folder_contains_target_hashtag_tweets = 'covid_tweet'

# the folder to store the collected small amount of tweets of the pretended target hashtags which are described in section 5.2 of the report
folder_contains_pretended_target_hashtag_tweets = 'evaluation_testing_files/testing_tweets_data'
os.chdir(main_folder)
```

```python
files = [file for file in os.listdir(folder_contains_target_hashtag_tweets) if "tag_#" in file]
files
```

```python
import datetime

for file in files:
  df = pd.read_csv('./covid_tweet/' + file, lineterminator='\n')
  df["date"] = pd.to_datetime(df["date"], errors = "coerce").dt.date
  df = df[(df['date'] > datetime.date(year=2021,month=1,day=30))]
  df.to_csv('./' + folder_contains_pretended_target_hashtag_tweets + '/eval_' + file)
```
